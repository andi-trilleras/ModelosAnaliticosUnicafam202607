##Listas


```python
#Una lista vacia
mi_lista=[]

#lista con varios elementos
mi_lista=[1,2,3,"4",5.0,True]
print(mi_lista[2])
```

```python
#Cambios en la lista
mi_lista[0]="Andrea"
print(mi_lista)
```

```python
#Agregar elementos a una lista
mi_lista.append("Trilleras")
print(mi_lista)
```

```python
#Agregar elementos a una posicion especifica
mi_lista.insert(3,"Bogota")
print(mi_lista)
```

```python
#Agregar varios elementos al final de la lista
mi_lista.extend([7,False])
print(mi_lista)
```

```python
#Quitar elementos
#Eliminar objetos especificos
mi_lista.remove(False)
print(mi_lista)
```

```python
print("ACTUALIZACION DE DATOS")
valor_buscar=input("Por favor ingrese el nombre para actualizar")
valor_reemplazar=input("Ingrese valor a reemplazar")
mi_lista[mi_lista.index(valor_buscar)]=valor_reemplazar
print("Cambio ejecutado, imprimiendo resultado")
print(mi_lista)
```

##Diccionarios

```python
mi_diccionario={
    "Nombre":"Andrea",
    "Ciudad de nacimiento": "Bogota",
    "Empresa":"Omnicom"
}
```

```python
#Acceder a elementos del diccionario
print(mi_diccionario["Nombre"])
```

```python
#Modificar un valor en un diccionario
mi_diccionario["Nombre"]="Andrea T"
print(mi_diccionario)
```

```python
#Añadir un elemento al diccionario
mi_diccionario["Profesion"]="Ing. Industrial"
print(mi_diccionario)
```

```python
#Eliminar un elemento del diccionario
del mi_diccionario ["Ciudad de nacimiento"]
print(mi_diccionario)
```

##Condicionales

```python
#Crea un programa que determine si la persona es mayor o menor de edad
print("----------VERIFICACION DE EDAD------------")
edad=int(input("Ingresa tu edad:"))
if edad<18:
  print(f"Eres menor de edad, tienes:{edad}")
else:
  print(f"Eres mayor de edad, tienes:{edad}")
```

```python
#Condicionales múltiples
nota=85
if nota>=80:
  if nota>=90:
    print("Excelente")
  elif nota>=80 and nota<90:
    print("Aceptable")
else:
  print ("Insuficiente")
```
