```python
#Dado peso = 70 (kg) y altura = 1.75 (m), calcula el IMC (peso / altura²) e imprime la categoría: Bajo peso (<18.5), Normal (18.5–24.9), Sobrepeso (25–29.9), Obesidad (≥30).
peso=70
altura=1.75
imc=peso/altura**2
if (imc>=30):
  clasificacion="Obesidad"
elif (imc>=25):
  clasificacion="Sobrepeso"
elif (imc>=18.5):
  clasificacion="Normal"
else:
  clasificacion="Bajo peso"
print(clasificacion)
```

```python
#Una tienda aplica descuentos según el monto: 0% si es menos de $50,000; 10% entre $50,000 y $99,999; 20% entre $100,000 y $199,999; 30% si es $200,000 o más. Dado monto = 120000, calcula e imprime el descuento y el total a pagar.
monto = 120000

if monto < 50000:
    dcto = 0
elif monto < 100000:
    dcto = 0.10
elif monto < 200000:
    dcto = 0.20
else:
    dcto = 0.30

dcto_total = monto * dcto
total  = monto - dcto_total
print(f"Descuento: {int(dcto*100)}% (${dcto_total})")
print(f"Total: ${total}")
```
