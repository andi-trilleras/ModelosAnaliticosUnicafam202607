##Ejercicios de Valores de Cadena

```python
##Declara una cadena llamada saludo con el valor "Hola, mundo!"
saludo="Hola, mundo!"
print(saludo)
```

```python
##Accede y muestra el primer carácter de la cadena saludo
print(saludo[0])
```

```python
##Accede y muestra los últimos 5 caracteres de la cadena saludo
print(saludo[7:12])
```

```python
#Concatenar la cadena saludo con otra cadena ", ¿cómo estás?" y asígnala a una nueva variable saludo_completo.
saludo2="¿cómo estás?"
saludo_completo= saludo+saludo2
print(saludo_completo)
```

```python
#Convierte la cadena saludo_completo a mayúsculas y asígnala a una variable saludo_mayusculas
saludo_mayusculas= saludo_completo.upper()
print(saludo_mayusculas)
```

```python
#Reemplaza la palabra "mundo" en saludo_completo por "Python" y asígnala a una variable saludo_reemplazo.
saludo_reemplazo = saludo_completo.replace("mundo", "Python")
print(saludo_reemplazo)
```
