Crea un programa que verifique si una persona es mayor de edad. Una persona es considerada mayor de edad si tiene 18 años o más.

```python
print("VERIFICACION DE EDAD")
edad=int(input("Ingresa tu edad:"))
if edad<18:
  print(f"Eres menor de edad, tienes:{edad}")
else:
  print(f"Eres mayor de edad, tienes:{edad}")
```

Escribe un programa que verifique si un número ingresado por el usuario es par o impar.

```python
print("VERIFICACION PAR O IMPAR")
numero=int(input("Ingresa el numero:"))
if numero%2==0:
  print(f"El numero es par")
else:
  print(f"El numero es impar")
```

Escribe un programa que clasifique la edad en diferentes categorías: Niño (0-12 años), Adolescente (13-17 años), Adulto (18-64 años) y Anciano (65 años o más).

```python
print("Clasificacion de edades")
edad2=int(input("Ingresa la edad:"))
if edad2>=65:
  print(f"La categoria es: Anciano")
elif edad2>=18:
  print(f"La categoria es: Adulto")
elif edad2>=13:
  print(f"La categoria es: Adolescente")
else:
  print(f"La categoria es: Niño")
```

Crea un programa que lea dos números y determine cuál es mayor, cuál es menor o si ambos son iguales.

```python
print("Identificar números")
n1=int(input("Ingresa el primer numero:"))
n2=int(input("Ingresa el segundo numero:"))
if n1>n2:
  print(f"El numero mayor es:{n1}")
elif n2>n1:
  print(f"El numero mayor es:{n2}")
else:
  print(f"Los números son iguales")
```
